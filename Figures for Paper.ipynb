```python
import qutip.testing
import numpy as np
import matplotlib.pyplot as plt
import math
#qutip.testing.run()

```

```python
b = qutip.Bloch()
b.make_sphere()
vec = [[0.000, 1, 0], [0.707, 0.707, 0.0000] ]
b.add_vectors(vec)
b.render()
plt.savefig('BB84BasicStates.png')
```

```python
triX = 3/(2*math.sqrt(3))
triY = -1/2

b.fig.clf()
vec = [0, 1, 0]
b.add_vectors(vec)

b.fig.clf()
vec1 = [-triX, triY, 0]
b.add_vectors(vec1)

b.fig.clf()
vec2 = [triX, triY, 0]
b.add_vectors(vec2)
b.render()
```
